```python
k=4
dynamic_aux = [-1]*40
dynamic_aux[0] = dynamic_aux[1] = 1

def WascallyWabbits(n):
    global dynamic_aux,k
    if dynamic_aux[n] == -1:
        dynamic_aux[n] = WascallyWabbits(n-1) + WascallyWabbits(n-2)*k
    return dynamic_aux[n]

WascallyWabbits(32)
```

```python
from Bio import SeqIO
records = list(SeqIO.parse("Computing_GC_Content.fasta", "fasta"))
max = 0
for record in records:    
    count = 0
    count_gc = 0
    for letter in record.seq:
        if letter == 'G' or letter == 'C':
            count_gc +=1
        count+=1
    result = (count_gc/count)*100
    #print("Result: %.6f Count: %d, Count_gc: %d"%(result,count,count_gc))
    if result > max:
        max = result 
        final = record

print(final.id)
print("%.6f"%(max))
```
